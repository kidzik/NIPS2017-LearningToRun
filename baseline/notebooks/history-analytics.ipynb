```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_pickle("../logs/rpm.pickle")
```

```python
obs = [data[0][i][0] for i in range(len(data[0]))]
```

```python
obs_arr = np.array(obs)
desc = stats.describe(obs_arr)
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (10,10)

plt.bar(range(len(desc.mean)), desc.minmax[1])
plt.bar(range(len(desc.mean)), desc.minmax[0])
plt.show()

plt.bar(range(len(desc.mean)), desc.variance)
plt.show()
```
